```python
import re

def clean_format(text):
    text  =text.replace('..','.')
    text  =text.replace('..','.')
    text  =text.replace(',.','.')
    text  =text.replace('.,','.')
    text  =text.replace(', .','.')
    text  =text.replace('. ,','.')
    text  =text.replace('.',' . ')
    text  =text.replace(',',' , ')
    text = text.replace('మరియు',",")
    text = text.replace('యొక్క'," ")
    text = text.replace(" . ",". ")
    text = text.replace(" , ",", ")
    text = text.replace(' , ',", ")
    text = text.replace(' . ',". ")
    text = text.replace('  '," ")
    text = text.replace('  '," ")
    text = text.replace(',,',",")
    text = text.replace(',,',",")
    text = text.replace(',.',".")
    text = text.replace('.,',".")
    text = text.replace('. ,','.')
    text = text.replace(', .',".")
    text = text.replace(' . ',". ")
    text = text.replace(' . ',". ")
    text = text.replace(' . ',". ")
    text = text.replace(' . ',". ")
    text = text.replace(' , ',", ")
    text = text.replace(' , ',", ")
    text = text.replace(' , ',", ")
    text = text.replace(' , ',", ")
    text = text.replace('  '," ")
    text = text.replace('  '," ")
    text = text.replace('  '," ")
    text = text.replace(" . ",". ")
    text = text.replace(" , ",", ")
    text = text.replace(' , ',", ")
    text = text.replace(' . ',". ")
    text = text.replace(',,',",")
    text = text.replace(',,',",")
    text = text.replace(',.',".")
    text = text.replace('.,',".")
    text = text.replace('  '," ")
    text = text.replace('  '," ")
    text = text.replace('  '," ")
    return text

def clean_tel_num(text):
    tel_num = { '౦':'0','౧':'1','౨':'2','౩':'3','౪':'4','౫':'5','౬':'6','౭':'7','౮':'8','౯':'9'}
    for k,v in enumerate(tel_num):
        text.replace(v,str(k))
    return text

def clean_des(text):
    if pd.isna(text):
        return np.nan
    text = text.strip()
    text = text.strip('"')
    if text.startswith("'''"):
        pass
    elif text.startswith("''''"):
        pass
    elif text.startswith("'''"):
        text = "'"+text
    elif text.startswith("''"):
        text = "'"+text
    if not text.endswith('.'):
        text = text + '.'
    text = clean_tel_num(text)
    text = clean_format(text)
    
    lis_t = text.split("'''")
    
    try:
        while(1):
            lis_t.remove('')
    except ValueError:
        pass
    
    if len(lis_t)%2:
        pass
    else:
        flag = 0
        t = ''
        for p in lis_t:
            if flag:
                t += " ===\n" + p
                flag = 0
            else:
                t += "=== " + p
                flag = 1     
        text = t

    text = text.replace('+++++',"'''''")
    text = text.replace('****',"''''")
    if "''''" in text:
        lis_2 = text.replace('\n','').split("''''")
        try:
            while(1):
                lis_2.remove('')
        except ValueError:
            pass

        if len(lis_t)%2:
            pass
        else:
            flag = 0
            t = ''
            for p in lis_t:
                if flag:
                    t += " ''''\n" + p
                    flag = 0
                else:
                    t += "'''' " + p
                    flag = 1     
            text = t
            

    return text


lis_cols = list(awards.columns)
for i in lis_cols:
    if 'Telugu' in i and i != 'Recipients Telugu':
        awards[i] = awards[i].apply(clean_des)

out('_v2',awards,True)
        
```

```python
awards.set_index('Recipients',inplace=True)
def getData(data,A):
    row_dict = {}
    award_dict = dict(data.loc[A])
    # print(award_dict)
    for k,v in zip(award_dict,award_dict.values()):
        if pd.isna(v) :
            row_dict["_".join(k.split(' '))] = 'NaN'
        else:
            row_dict["_".join(k.split(' '))] = str(v).strip()
    numericals = ['Padma_Shri_Year','Padma_Bhushan_Year','Padma_Vibhushan_Year','Page_Id']
    new_dic = {}
    flag = 0
    remove_list = ['Summary_Telugu','further_reading_Telugu']
    for k,v in row_dict.items():
        if k == 'Summary':
            flag = 1
        if flag and 'Telugu' in k and k not in remove_list:
            new_dic[k] = v
    row_dict['body'] = new_dic
    for c_n in numericals:
        # print(c_n)
        if row_dict[c_n] != 'NaN':
            try:
                row_dict[c_n] = str(int(float(row_dict[c_n])))
            except:
                continue
    row_dict['Recipients'] = A
    ref_dic = {}
    if row_dict['Ref'] != 'NaN':
        refs =  row_dict['Ref'].split(';')
        for ref in refs:
            ref_dic[ref] = ref.split('//')[1].split('/')[0]
        print(ref_dic)
        row_dict['ref_dict'] = ref_dic
    return row_dict
```
