```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup 

def infobox_scraper(url):
  # table_class="wikitable sortable jquery-tablesorter"
  response=requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  table=soup.find('table',{'class':"infobox biography vcard"})
  df=pd.read_html(str(table))
  df=pd.DataFrame(df[0]) 
  df=df.T
  res=[]
  for column in df.columns:
    li = df[column].tolist()
    res.append(li)
  return res
```

```python
awards = pd.read_excel('/content/renders_awards_v1_03_01_awards_Kaggle_out.xlsx')
import numpy as np
from tqdm import tqdm
# df_main = pd.DataFrame()
# df_1=pd.DataFrame()
final=[]
empty=[]
awards.reset_index(inplace=True)
for i in tqdm(range(awards.shape[0])):
    k = awards['WikiLink'][i]
    # print(k)
    if pd.isna(k):
        final.append(empty)
    else:
      try:
        # print(infobox_scraper(k))
        df_trial=(infobox_scraper(k))
        final.append(df_trial)
        print("added to main list")
      except:
        print("An exception occurred")

print(final)
df_main = pd.DataFrame(final)
```

```python
file_name = 'infobox_main.xlsx'

(df_main).to_excel(file_name)
print('DataFrame is written to Excel File successfully.')
```
