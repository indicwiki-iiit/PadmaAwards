```python
import re
import datefinder

def get_date(string_with_dates):
    matches = datefinder.find_dates(string_with_dates)
    for match in list(matches):
        return str(match.date())
    if len(list(matches)) == 0 :
        return np.nan

def get_year(text):
    k = re.findall('\d{4}',text)
    if len(k) == 0:
        return np.nan
    return k[0]

for i in tqdm(range(awards.shape[0])):
    b = awards['Died'][i]
    if not pd.isna(b):
        b = b.replace('\n',' ').replace('\t',' ')
        d = get_date(b)
        awards['Death Date'][i] = d

for i in tqdm(range(awards.shape[0])):
    d2 = awards['Death Date'][i]
    if pd.isna(d2):
        b = awards['Died'][i]
        if not pd.isna(b):
            b = b.replace('\n',' ').replace('\t',' ')
            d = get_year(b)
            awards['Death Date'][i] = d
```
